```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "Method": [
        "All-at-once", "Step-by-step",    # English-1
        "All-at-once", "Step-by-step",    # English-2
    ],
    "Dataset": [
        "Clinical-English", "Clinical-English",
        "PARROT-English", "PARROT-English",
    ],
    "Value": [
        0.806, 0.819,   # English-1
        0.768, 0.792,   # English-2
    ]
}

df = pd.DataFrame(data)

sns.set_style("whitegrid")

# Create the grouped bar chart using seaborn
plt.figure(figsize=(10, 7)) 
sns.barplot(x="Method", y="Value", hue="Dataset", data=df, palette="viridis", width=0.4)

plt.title("Mean semantic similarity per method", fontsize=16)
plt.xlabel("Filling method", fontsize=12)
plt.ylabel("Mean semantic similarity", fontsize=12)

# Set the y-axis limit from 0 to 1 as these are metric scores
plt.ylim(0, 1)

plt.legend(title="Dataset", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()

plt.show()

## Evaluation of report filling methods: step-by-step vs. all-at-once. Across all datasets, step-by-step consistently outperforms in both semantic similarity and percentage of superior outputs.
```
